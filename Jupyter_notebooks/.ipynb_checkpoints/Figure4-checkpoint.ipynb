# Biases conditional on WBGT values

- It is useful to know whether the biases of sWBGT and ESI are independent of WBGT values or not. A correlation between them indicates biases of different magnitudes for heat stress of different levels. 
- To explore it, we compare biases conditional on the 50th, 75th, 90th, 95th, 99th, and 99.9th percentile exceedance values of WBGT which is done for each individual year first and then averaged across the period 1990-2019.  
- In this notebook, we will reproduce Figure 4 in "Explicit Calculations of Wet Bulb Globe Temperature compared with approximations and why it matters for labor productivity" by Qinqin Kong and Matthew Huber. 

````python
import xarray as xr
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import gridspec
import cmaps
import regionmask
import cartopy.crs as ccrs
````

````python
%matplotlib inline
%config InlineBackend.figure_format = "retina"
````

### Read in data:
- We read in the following file:
  - ```conditional_bias.nc``` which contains:
      - sWBGT biases conditional different quantiles of WBGT: ```swbgt_bias```
      - ESI biases conditional different quantiles of WBGT: ```esi_bias```


- It was derived from hourly ERA5 reanalysis data covering a 30-year period of 1990-2019.

- We will use it to explore the potential correlation between biases and WBGT values.

````python
# read in data
conditional_bias=xr.open_dataset('../nc_files/conditional_bias.nc')
conditional_bias
````

### create land mask

We use ```regionmask``` python package to create a land mask which enables us to plot data only over land.

````python
mask = regionmask.defined_regions.natural_earth.land_110.mask(conditional_bias,lon_name="longitude",lat_name="latitude")
mask.plot()
````

### Reproduce Figure 4

````python
# create meshgrid of latitude and longitude
LON,LAT=np.meshgrid(conditional_bias.longitude,conditional_bias.latitude)  

# function to plot each individual subplot
def subplot(data,axis,norm,cmap,landmask):
    g=axis.pcolormesh(LON, LAT,data.where(landmask==0),transform=ccrs.PlateCarree(),cmap=cmap,norm=norm,rasterized=True) # make the plot
    axis.coastlines(linewidth=0.5) # add coastlines
    axis.set_title('')
    axis.set_extent([-169.999,189.999,-60,70]) # set extent of the map to exclude high latitudes
    return g

# function to make the whole plot    
def plot(conditional_bias):
    # create figure
    fig = plt.figure(figsize=[12, 7])
    # add axis
    ax1 = fig.add_subplot(4, 3, 1, projection=ccrs.Mollweide(central_longitude=10))
    ax2 = fig.add_subplot(4, 3, 2, projection=ccrs.Mollweide(central_longitude=10))
    ax3 = fig.add_subplot(4, 3, 3, projection=ccrs.Mollweide(central_longitude=10))
    ax4 = fig.add_subplot(4, 3, 4, projection=ccrs.Mollweide(central_longitude=10))
    ax5 = fig.add_subplot(4, 3, 5, projection=ccrs.Mollweide(central_longitude=10))
    ax6 = fig.add_subplot(4, 3, 6, projection=ccrs.Mollweide(central_longitude=10))
    ax7 = fig.add_subplot(4, 3, 7, projection=ccrs.Mollweide(central_longitude=10))
    ax8 = fig.add_subplot(4, 3, 8, projection=ccrs.Mollweide(central_longitude=10))
    ax9 = fig.add_subplot(4, 3, 9, projection=ccrs.Mollweide(central_longitude=10))
    ax10 = fig.add_subplot(4, 3, 10, projection=ccrs.Mollweide(central_longitude=10))
    ax11 = fig.add_subplot(4, 3, 11, projection=ccrs.Mollweide(central_longitude=10))
    ax12 = fig.add_subplot(4, 3, 12, projection=ccrs.Mollweide(central_longitude=10))
    
    # adjust the relative position of subplots
    fig.subplots_adjust(bottom=0, left=0,right=1, top=1,hspace=0.15,wspace=-0.2)
    
    # add annotates
    anno_opts = dict(xy=(0, 1.05), xycoords='axes fraction',va='center', ha='center')
    num=['a','b','c','d','e','f','g','h','i','j','k','l']
    i=0
    for ax in [ax1,ax2,ax3,ax4,ax5,ax6,ax7,ax8,ax9,ax10,ax11,ax12]:
        ax.annotate(num[i], **anno_opts,fontsize=16)
        i+=1
    
    # set the colorbar scale
    levels = [-8, -7, -6, -5, -4, -3, -2, -1,  1,  2,  3,  4,  5,  6,  7,  8]
    norm=mpl.colors.BoundaryNorm(levels,ncolors=len(levels)+1,extend='both')
    
    # we pick colors from different colormaps, create a new color list, and then create a customized colormap from this new color list
    color1=cmaps.hotcolr_19lev.colors
    color2=cmaps.StepSeq25.colors[::-1]
    color3=cmaps.amwg_blueyellowred.colors
    color_new=np.concatenate((color1[0:8],np.atleast_2d(color1[9]),np.concatenate((color2[10:13],color3[9:12],color3[13:]))))
    cmap=mpl.colors.ListedColormap(color_new)
    
    # call subplot function
    g=subplot(conditional_bias.swbgt_bias.sel(quantiles=0.5),ax1,norm,cmap,mask)
    subplot(conditional_bias.swbgt_bias.sel(quantiles=0.75),ax2,norm,cmap,mask)
    subplot(conditional_bias.swbgt_bias.sel(quantiles=0.9),ax3,norm,cmap,mask)
    subplot(conditional_bias.swbgt_bias.sel(quantiles=0.95),ax4,norm,cmap,mask)
    subplot(conditional_bias.swbgt_bias.sel(quantiles=0.99),ax5,norm,cmap,mask)
    subplot(conditional_bias.swbgt_bias.sel(quantiles=0.999),ax6,norm,cmap,mask)
    subplot(conditional_bias.esi_bias.sel(quantiles=0.5),ax7,norm,cmap,mask)
    subplot(conditional_bias.esi_bias.sel(quantiles=0.75),ax8,norm,cmap,mask)
    subplot(conditional_bias.esi_bias.sel(quantiles=0.9),ax9,norm,cmap,mask)
    subplot(conditional_bias.esi_bias.sel(quantiles=0.95),ax10,norm,cmap,mask)
    subplot(conditional_bias.esi_bias.sel(quantiles=0.99),ax11,norm,cmap,mask)
    subplot(conditional_bias.esi_bias.sel(quantiles=0.999),ax12,norm,cmap,mask)
    
    # make color bar
    cb=fig.colorbar(g,ax=(ax1,ax2,ax3,ax4,ax5,ax6,ax7,ax8,ax9,ax10,ax11,ax12),orientation='horizontal', pad=0.06,shrink=0.6,aspect=40,spacing='proportional',
                    ticks=[-8.,-7.,-6., -5., -4., -3., -2., -1.,  1.,  2.,  3.,  4.,  5.,  6.,7.,8.]) 
    cb.set_label(label="sWBGT/ESI - WBGT ($^\circ$C)", size=14)
    cb.ax.tick_params(labelsize=12)
````

````python
plot(conditional_bias)
````

### Figure caption
Biases of sWBGT (a-f) and ESI (g-l) conditional on the 50th (a, g), 75th (b, h), 90th (c, i), 95th (d, j), 99th (e, k), and 99.9th (f, l) percentile exceedance values of WBGT during 1990-2019

# Findings
- Both sWBGT and ESI show a clear tendency towards smaller positive or stronger negative biases moving from lower to higher percentile exceedance values of WBGT, suggesting a potential correlation between biases and WBGT.

- sWBGT conditional on 50th percentile of WBGT shows substantial positive biases (>3°C globally) which are reduced to <2°C in the majority of the world when conditional on 99th percentile of WBGT. Negative biases even occur in many areas when we move to 99.9th percentile. 

- ESI exhibits small biases worldwide conditional on 50th percentile of WBGT which monotonically shift to strong negative biases conditional on 99.9th percentile of WBGT (<-1°C globally and <-4°C in the low latitudes).

- The dependence of biases on WBGT may be explained by the fact that both higher WBGT and stronger negative (or smaller positive) bias tend to be associated with strong solar radiation and light wind. 

- We expect sWBGT to largely overestimate median-level heat stress but less (or even underestimate) for more severe heat stress (such as the hottest week or 3 days of the year). ESI, in contrast, does a better job in measuring heat stress of median level but tend to seriously underestimate those of more severity.
