# Does temporal resolution matter for the estimation of annual labor productivity?
- In Figure7, annual labor productivity was estimated from hourly data which however may be not available in some archives such as CMIP and CORDEX. It is common to see studies using sub-daily, daily or even monthly output to estimate labor productivity. But the potential error introduced thereby hasn't been quantifed. 

- This notebook will estimate annual labor productivity from dataset of different temporal resolutions and compare them to get an idea of the error introduced by using data with coaser temporal resolution. Specifically, hourly ERA5 reanalysis data are re-sampled to 8 and 4 times daily scale (calculate temporal averages of radiation flux and re-sample instantaneous values of other fields once each 3 and 6 hours interval), and averaged to obtain the daily mean values. Then we estimate annual labor productivity with inputs of different temporal resolutions. Only Liljegren's WBGT formulation is used here.

- In this notebook, we will reproduce Figure S4 in "Explicit Calculations of Wet Bulb Globe Temperature compared with approximations and why it matters for labor productivity" by Qinqin Kong and Matthew Huber.

````python
import xarray as xr
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import gridspec
import cmaps
import regionmask
import cartopy.crs as ccrs
````

````python
%matplotlib inline
%config InlineBackend.figure_format = "retina"
````

### Read in data:
We read in the following files:
- ```labor_productivity_from_different_temporal_resolutions.nc``` which contains:
     - ```hourly```: annual average labor productivity calculated from hourly inputs
     - ```3hourly```: annual average labor productivity calculated from 8-times daily inputs
     - ```6hourly```: annual average labor productivity calculated from ESI 4-times daily inputs
     - ```daily```: annual average labor productivity calculated from daily average inputs
- ```labor_productivity_time_series_from_different_temporal_resolutions.nc``` which contains:
     - ```hourly```: times series of population weighted annual average labor productivity calculated from hourly inputs
     - ```3hourly```: times series of population weighted annual average labor productivity calculated from 8-times daily inputs
     - ```6hourly```: times series of population weighted annual average labor productivity calculated from 4-times inputs
     - ```daily```: times series of population weighted annual average labor productivity calculated from daily average inputs

````python
labor=xr.open_dataset('../nc_files/labor_productivity_from_different_temporal_resolutions.nc')
labor
````

````python
labor_ts=xr.open_dataset('../nc_files/labor_productivity_time_series_from_different_temporal_resolutions.nc')
labor_ts
````

### create land mask

We use ```regionmask``` python package to create a land mask which enables us to plot data only over land.

````python
mask = regionmask.defined_regions.natural_earth.land_110.mask(labor,lon_name="longitude",lat_name="latitude")
mask.plot()
````

````python
# create meshgrid of latitude and longitude
LON,LAT=np.meshgrid(labor.longitude,labor.latitude)
def subplot(data,axis,norm,cmap,landmask):
    g=axis.pcolormesh(LON, LAT,data.where(landmask==0)*100,transform=ccrs.PlateCarree(),cmap=cmap,norm=norm,rasterized=True)
    axis.coastlines(linewidth=0.5)
    axis.set_title('')
    axis.set_extent([-169.999,189.999,-60,70])
    return g
    
# function to make the whole plot    
def plot(labor,labor_ts):    
    
    # set the colorbar scale
    levels1 = [60,65,70,75,80,85,90,95,100.01]
    norm1=mpl.colors.BoundaryNorm(levels1,ncolors=len(levels1)-1,extend='neither')
    
    # we pick colors from different colormaps, create a new color list, and then create a customized colormap from this new color list
    color=cmaps.precip2_17lev.colors
    color_new=[color[14],color[13],color[12],color[10],color[9],color[4],color[2],color[0]]
    cmap1=mpl.colors.ListedColormap(color_new)
    
    # set the colorbar scale
    levels2 = [-13, -11, -9, -7, -5, -3, -1,  1,  3,  5,  7,  9,  11,  13]
    norm2=mpl.colors.BoundaryNorm(levels2,ncolors=len(levels2)+1,extend='both')
    
    # we pick colors from different colormaps, create a new color list, and then create a customized colormap from this new color list
    color1=cmaps.hotcold_18lev.colors
    color2=cmaps.StepSeq25.colors[::-1]
    color3= cmaps.amwg_blueyellowred.colors
    color_new=np.concatenate((color1[0:7],np.atleast_2d(color1[9]),color2[10:12],color3[9:]))
    cmap2=mpl.colors.ListedColormap(color_new)
    
    
    # use gridspec to customize the layout of subplots
    gs_top = plt.GridSpec(4, 3, top=0.95,hspace=0.3,width_ratios=[2,1,1],wspace=-0.6)
    gs_base = plt.GridSpec(4, 3,hspace=-0.1,wspace=0.4,width_ratios=[3.8,1.7,0.4],height_ratios=[1.5,2,2,2])
    fig = plt.figure(figsize=[15, 10])
    ax1 = fig.add_subplot(gs_top[0,0],projection=ccrs.Mollweide(central_longitude=10))
    ax2 = fig.add_subplot(gs_base[0,1])
    ax3 = fig.add_subplot(gs_top[1,0],projection=ccrs.Mollweide(central_longitude=10))
    ax4 = fig.add_subplot(gs_top[1,1:3],projection=ccrs.Mollweide(central_longitude=10))
    ax5 = fig.add_subplot(gs_top[2,0], projection=ccrs.Mollweide(central_longitude=10))
    ax6 = fig.add_subplot(gs_top[2,1:3], projection=ccrs.Mollweide(central_longitude=10))
    ax7 = fig.add_subplot(gs_top[3,0], projection=ccrs.Mollweide(central_longitude=10))
    ax8 = fig.add_subplot(gs_top[3,1:3], projection=ccrs.Mollweide(central_longitude=10))
    
    # adjust the relative position of subplots
    fig.subplots_adjust(bottom=0, left=0,right=1, top=1,hspace=0.6,wspace=0.6)
    
    # add annotates
    anno_opts = dict(xy=(0, 1.05), xycoords='axes fraction',va='center', ha='center')
    num=['a','b','c','d','e','f','g','h']
    i=0
    for ax in [ax1,ax2,ax3,ax4,ax5,ax6,ax7,ax8]:
        ax.annotate(num[i], **anno_opts,fontsize=16)
        i+=1
        
    # plot annual labor productivity calculated from inputs of each temporal resolution  
    g=subplot(labor['hourly'],ax1,norm1,cmap1,mask)
    subplot(labor['3hourly'],ax3,norm1,cmap1,mask)
    subplot(labor['6hourly'],ax5,norm1,cmap1,mask)
    subplot(labor['daily'],ax7,norm1,cmap1,mask)
    # make color bar
    cb1=fig.colorbar(g,ax=(ax1,ax3,ax5,ax7),orientation='horizontal', pad=0.04,shrink=0.3,
                     ticks=[60, 65, 70, 75,  80,  85,  90,  95,100],spacing='proportional',aspect=30)
    cb1.set_label(label="Labor capacity (%)", size=12)
    cb1.ax.tick_params(labelsize=12)
    
    # plot the anomaly of labor productivity introduced by using inputs of coarser temporal resolution
    h=subplot((labor['3hourly']-labor['hourly']),ax4,norm2,cmap2,mask)
    subplot((labor['6hourly']-labor['hourly']),ax6,norm2,cmap2,mask)
    subplot((labor['daily']-labor['hourly']),ax8,norm2,cmap2,mask)
    
    # plot time series of population weighted annual labor productivity calculated from inputs of each temporal resolution
    ax2.plot(labor_ts.year,labor_ts['hourly']*100,label='hourly',color='red')
    ax2.plot(labor_ts.year,labor_ts['3hourly']*100,label='3-hourly',color='orange')
    ax2.plot(labor_ts.year,labor_ts['6hourly']*100,label='6-hourly',color='blue')
    ax2.plot(labor_ts.year,labor_ts['daily']*100,label='daily',color='green')
    ax2.set_ylim(90, 100)
    ax2.yaxis.set_ticks_position('right')
    ax2.yaxis.set_label_position('right')
    ax2.set_ylabel('Labor capacity (%)',fontsize=12)
    ax2.set_xlabel('')
    ax2.legend()
    ax2.tick_params(axis='both', which='major', labelsize=12)
    
    # make color bar
    cb2=fig.colorbar(h,ax=(ax2,ax4,ax6,ax8),orientation='horizontal', pad=0.04,aspect=30,shrink=0.5,
                     ticks=[-13, -9, -5, -1,  1,  5, 9, 13],spacing='proportional') 
    cb2.set_label(label="Labor capacity anomaly (%)", size=14)
    cb2.ax.tick_params(labelsize=12)
````

````python
plot(labor,labor_ts)
````

## Figure caption
Annual average labor productivity calculated from hourly inputs (a); annual average labor productivity calculated from 8 times daily (c), 4 times daily (e), and daily average inputs (g), and their anomalies (d, f, h respectively) compared with that based on hourly inputs (other resolutions minus hourly); population weighted global annual average labor productivity during 1950-2019 (b).

# Findings
- Labor productivity derived from daily average inputs is substantially overestimated especially in the tropics (by around 7 to more than 13 percent), which is not surprising since both WBGT formulation and labor productivity function are nonlinear. Particularly, all existing labor productivity functions involve a lower threshold of WBGT (e.g. 25°C for very heavy work with a metabolic rate of 520W according to ISO7243 standard) below which there is no labor loss. It is likely to have a daily average WBGT below this threshold but much higher WBGT values during peaking daytime hours in which case the labor productivity estimated from daily average WBGT is too optimistic. 

- In terms of population-weighted global and annual average labor productivity, the adoption of daily average inputs introduce a consistent overestimation by around 2.2 percent during the period 1950-2019. Nevertheless, the derived decreasing trend is similar between hourly (-0.33 percent per decade) and daily average inputs (-0.29 percent per decade). In comparison, the 8 or 4 times daily inputs mainly face a sampling issue (despite the time average for radiation fields) which nevertheless only introduce small errors of within $\pm1$ percent in most of the world.
