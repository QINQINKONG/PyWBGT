# Biases of extreme values

- Extreme events are of special importance in the study of heat stress. The stronger negative biases of both metrics conditional on higher percentile exceedance values of WBGT (see Figure 4) raise a cautionary note that extreme heat stress at some places of the world may be seriously underestimated. 

- We implemented a generalized extreme value (GEV) analysis to estimate and compare the extreme values of WBGT, sWBGT and ESI at each grid cell. A GEV model was fit to the annual maximum (calculated from hourly frequency) of each metric during 1990-2019 using ERA5 reanalysis data. The metric values corresponding to a 1-in-30-year event are calculated and compared.

- In this notebook, we will reproduce Figure 5 in "Explicit Calculations of Wet Bulb Globe Temperature compared with approximations and why it matters for labor productivity" by Qinqin Kong and Matthew Huber. 

````python
import xarray as xr
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import gridspec
import cmaps
import regionmask
import cartopy.crs as ccrs
````

````python
%matplotlib inline
%config InlineBackend.figure_format = "retina"
````

### Read in data:
- We read in the following file:
  - ```extreme_values.nc``` which contains:
      - 30-year return level of WBGT: ```wbgt```
      - 30-year return level of sWBGT: ```swbgt```
      - 30-year return level of ESI: ```esi```
      - differences between the 30-year return levels of sWBGT and WBGT (sWBGT-WBGT): ```swbgt_wbgt```
      - differences between the 30-year return levels of ESI and WBGT (ESI-WBGT): ```esi_wbgt```


- It was derived from hourly ERA5 reanalysis data covering a 30-year period of 1990-2019.

- We will use it to investigate biases of extreme values for both metrics.

````python
# read in data
gev=xr.open_dataset('../nc_files/extreme_values.nc')
gev
````

### create land mask

We use ```regionmask``` python package to create a land mask which enables us to plot data only over land.

````python
mask = regionmask.defined_regions.natural_earth.land_110.mask(gev,lon_name="lon",lat_name="lat")
mask.plot()
````

### Reproduce Figure 5

````python
# create meshgrid of latitude and longitude
LON,LAT=np.meshgrid(gev.lon,gev.lat)

# function to plot each individual subplot
def subplot(data,axis,norm,cmap):
    g=axis.pcolormesh(LON, LAT,data,transform=ccrs.PlateCarree(),cmap=cmap,norm=norm,rasterized=True) # make the plot
    axis.coastlines(linewidth=0.5) # add coastlines
    axis.set_title('')
    axis.set_extent([-169.999,189.999,-60,70]) # set extent of the map to exclude high latitudes
    return g
def plot(gev):    
    # set colorbar scale
    levels1 = [24,26,28,30,32,34,36,38,40,42]
    norm1=mpl.colors.BoundaryNorm(levels1,ncolors=len(levels1)+1,extend='both')
    # we pick colors from different colormaps, create a new color list, and then create a customized colormap from this new color list
    color=cmaps.sunshine_diff_12lev.colors
    color_new=[color[1],color[2],color[3],color[4],color[5],color[7],color[8],color[9],color[10],color[11],color[12]]
    cmap1=mpl.colors.ListedColormap(color_new)
    
    # set colorbar scale
    levels2 = [-8, -7, -6, -5, -4, -3, -2, -1,  1,  2,  3,  4,  5,  6,  7,  8]
    norm2=mpl.colors.BoundaryNorm(levels2,ncolors=len(levels2)+1,extend='both')
    # we pick colors from different colormaps, create a new color list, and then create a customized colormap from this new color list
    color1=cmaps.hotcolr_19lev.colors
    color2=cmaps.StepSeq25.colors[::-1]
    color3= cmaps.amwg_blueyellowred.colors
    color_new=np.concatenate((color1[0:8],np.atleast_2d(color1[9]),np.concatenate((color2[10:13],color3[9:12],color3[13:]))))
    cmap2=mpl.colors.ListedColormap(color_new)
    
    # create figure
    fig = plt.figure(figsize=[13, 12])
    # use gridspec to customize the layout of subplots
    spec = gridspec.GridSpec(ncols=6, nrows=2,width_ratios=[1,1,1,1,1,1], wspace=0.1,hspace=-1.05, height_ratios=[1,1])
    ax1 = fig.add_subplot(spec[0:2], projection=ccrs.Mollweide(central_longitude=10))
    ax2 = fig.add_subplot(spec[2:4], projection=ccrs.Mollweide(central_longitude=10))
    ax3 = fig.add_subplot(spec[4:6], projection=ccrs.Mollweide(central_longitude=10)) 
    ax4 = fig.add_subplot(spec[7:9], projection=ccrs.Mollweide(central_longitude=10))
    ax5 = fig.add_subplot(spec[9:11], projection=ccrs.Mollweide(central_longitude=10))
    
    # add annotates
    fig.subplots_adjust(bottom=0, left=0,right=1, top=1,wspace=0.03)
    anno_opts = dict(xy=(0, 1.05), xycoords='axes fraction',va='center', ha='center')
    num=['a','b','c','d','e']
    i=0
    for ax in [ax1,ax2,ax3,ax4,ax5]:
        ax.annotate(num[i], **anno_opts,fontsize=16)
        i+=1
    
    # call subplot function
    g=subplot(gev.wbgt-273.15,ax1,norm1,cmap1) # plot return values of wbgt
    subplot(gev.swbgt-273.15,ax2,norm1,cmap1) # plot return values of swbgt
    subplot(gev.esi,ax3,norm1,cmap1) # plot return values of esi
    
    # make colorbar
    cb1=fig.colorbar(g,ax=(ax1,ax2,ax3),orientation='horizontal', pad=0.01,shrink=0.5,fraction=0.5,aspect=30)
    cb1.set_label(label="WBGT, sWBGT or ESI ($^\circ$C)", size=14)
    cb1.ax.tick_params(labelsize=12)
    
    # call subplot function
    h=subplot(gev.swbgt_wbgt,ax4,norm2,cmap2) # plot differences between return values of swbgt and wbgt
    subplot(gev.esi_wbgt,ax5,norm2,cmap2)  # plot differences between return values of esi and wbgt
    
    # make color bar
    cb2=fig.colorbar(h,ax=(ax4,ax5),orientation='horizontal', pad=0.01,shrink=0.6,fraction=0.2,aspect=30,
                     ticks=[-7, -5, -3, -1,  1,  3,  5,  7],spacing='proportional') 
    cb2.set_label(label="sWBGT/ESI - WBGT ($^\circ$C)", size=14)
    cb2.ax.tick_params(labelsize=12)
    
````

````python
plot(gev)
````

### Figure caption
WBGT (a) and ESI (b) return levels corresponding to a 1-in-30-year event, and their differences (ESI-WBGT) (c)

# Findings

- Biases of extreme values share similar pattern with those conditional on 99.9 percentile exceedance values of WBGT (see Figure 4f,l) yet with larger magnitudes. 

- Even in extreme value sWBGT produces overestimated values in many regions with the notable exception of subtropical dry regions. Large negative biases are detected in North Africa and Middle East region (MENA) (-4°C to -7°C). 

- ESI underestimates WBGT by more than 3°C across most of the world. MENA regions stand out with strong negative biases between -6°C and -10°C.

- The biases structure of extreme values shown here is not captured at climatological mean level (see Figure 2). For example, relatively small biases of ESI at climatological mean level suggest it is a potentially acceptable approximation of WBGT for quantifying climatological mean heat stress or its temporal trends. Nevertheless, serious underestimations are expected when it comes to extreme heat stress.
