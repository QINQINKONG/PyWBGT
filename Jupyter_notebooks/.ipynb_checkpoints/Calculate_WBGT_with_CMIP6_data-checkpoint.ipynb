# Calculate WBGT with CMIP6 data
This notebook will introduce how to calculate WBGT with CMIP6 data in google cloud using the code developed by Qinqin Kong and Matthew Huber.

The code was transferred (with a bit modifications) from Liljegren's original code which is in C language (available from https://github.com/mdljts/wbgt/blob/master/src/wbgt.c). For details of the calculation procedure, please refer to Liljegren et al (2018).

# 1. import packages

### 1.1 import general packages

````python
# import general packages
import xarray as xr
import numpy as np
from matplotlib import pyplot as plt
from numba import njit, vectorize
# import packages needed for obtaining google cloud data
import pandas as pd
import fsspec
````

### 1.2 import modules for WBGT calculation
We import two modules:
- ```coszenith``` from which we import two functions:
  - ```cosza```: calculating cosine zenith angle during each interval
  - ```coszda```: calculating cosine zenith angle during only the sunlit part of each interval
- ```WBGT``` from which we import:
  - ```WBGT_Liljegren```: the original formulation in Liljegren's code. It only require surface downward solar radiation as radiation input, and other radiation components are approximated.
  - ```WBGT_GCM```: our modified formulation to directly use the full set of radiation components (surface downward and upwelling solar and thermal radiation) that are available from climate model output.

````python
# import functions for calculating cosine zenith angle
from coszenith import coszda, cosza
# import functions for calculating WBGT
from WBGT import WBGT_Liljegren, WBGT_GCM
````

# 2. Read in cmip6 data in google cloud
We will use cmip6 data in google cloud to illustrate how to calculate WBGT with our python code.

First, read in a csv files that contain the metadata information of cmip6 data that are available in google cloud

````python
df = pd.read_csv('https://storage.googleapis.com/cmip6/cmip6-zarr-consolidated-stores.csv')
````

Query the dataframe to get the variables that are needed for calculating WBGT. Here we select historical experiment from only one model (CNRM-CM6-1) at 3 hourly frequency.

````python
subset_df=df.query("source_id=='CNRM-CM6-1'& experiment_id=='historical' & table_id=='3hr' & variable_id==['tas', \
'huss','uas','vas', 'rsds', 'rsus', 'rlds', 'rlus', 'ps','rsdsdiff'] & member_id=='r1i1p1f2'")
subset_df
````

We define functions to read in data with xarray.

````python
def drop_all_bounds(ds):
    """Drop coordinates like 'time_bounds' from datasets,
    which can lead to issues when merging."""
    drop_vars = [vname for vname in ds.coords
                 if (('_bounds') in vname ) or ('_bnds') in vname]
    return ds.drop(drop_vars)
def open_dsets(df):
    """Open datasets from cloud storage and return xarray dataset."""
    dsets = [xr.open_zarr(fsspec.get_mapper(ds_url), consolidated=True)
             .pipe(drop_all_bounds)
             for ds_url in df.zstore]
    return dsets
````

read in data and store multiple xarray dataset within a dictionary

````python
dsets = dict() 
for group, df in subset_df.groupby(by=['variable_id']):
    dsets[group] = open_dsets(df)
````

Here we access each variable, and load them into memory. We just select one day (2001-07-01) as an example.

````python
huss=dsets['huss'][0].huss.sel(time='2001-07-01').load() #specific humidity
tas=dsets['tas'][0].tas.sel(time='2001-07-01').load() # 2m air temperature
ps=dsets['ps'][0].ps.sel(time='2001-07-01').load() # surface air pressure
uas=dsets['uas'][0].uas.sel(time='2001-07-01').load() # 10m u wind
vas=dsets['vas'][0].vas.sel(time='2001-07-01').load() # 10m v wind
rlds=dsets['rlds'][0].rlds.sel(time='2001-07-01').load() # long wave downwelling radiation
rsds=dsets['rsds'][0].rsds.sel(time='2001-07-01').load() # short wave downwelling solar radiation
rlus=dsets['rlus'][0].rlus.sel(time='2001-07-01').load() # long wave upwelling radiation
rsus=dsets['rsus'][0].rsus.sel(time='2001-07-01').load() # short wave upwelling solar radiation (reflected by ground)
rsdsdiff=dsets['rsdsdiff'][0].rsdsdiff.sel(time='2001-07-01').load() #downwelling diffused solar radiation
````

# 3. Data preprocessing

Before call the function to calculate WBGT, we need to do some preprocessing on the data.

### 3.1 Interpolate radiation
First, we notice that there is an offset between the time point of nonradiation fields (temperature, humidity and wind) and the center point of the interval across which radiation fluxes are averaged. In this case, temperature, humidity and wind fields are located at 00:00:00, 03:00:00, 06:00:00..., whereas radiation fields are centered on 01:30:00, 04:30:00, 07:30:00 ... Hence, we need to interpolate radiation fields to the same time points as other fields.

````python
# function to interpolate radiation fields
def interp(data):
    date=np.arange(np.datetime64('2001-07-01T03:00:00'), np.datetime64('2001-07-01T23:00:00'),np.timedelta64(3, 'h'))
    tmp=data.interp(time=date,method='linear')
    result=xr.concat([data.isel(time=0).assign_coords(time=np.datetime64('2001-07-01T00:00:00')),tmp],dim='time').transpose("time", "lat", "lon")

    return result
````

````python
# interpolate radiation
rsdsinterp=interp(rsds)
rsusinterp=interp(rsus)
rldsinterp=interp(rlds)
rlusinterp=interp(rlus)
rsdsdiffinterp=interp(rsdsdiff)
````

````python
# check that radiation fields have correct time coordinates
rsdsinterp.time
````

### 3.2 calculate cosine zenith angle
We calculate cosine zenith angle averaged both during each interval and during only the sunlit period of each interval. Both of them are needed for calculating WBGT. For the details of the calculation procedure of cosine zenith angle please refer to Hogan and Hirahara (2016), and Di Napoli et al (2020)

````python
# create meshgrid of latitude and longitude, and we will calculate cosine zenith angle on these grids
lon,lat=np.meshgrid(huss.lon,huss.lat)
lat=lat*np.pi/180
lon=lon*np.pi/180
````

````python
#calculate cosine zenith angle averaged during each interval
cza=cosza(huss.time.values,lat,lon,3)
cza=xr.DataArray(cza,dims=("time", "lat",'lon'),coords={"time":huss.time,"lat":huss.lat,"lon":huss.lon})

#calculate cosine zenith angle averaged during only the sunlit part of each interval
czda=coszda(huss.time.values,lat,lon,3)
czda=xr.DataArray(czda,dims=("time", "lat",'lon'),coords={"time":huss.time,"lat":huss.lat,"lon":huss.lon})
# here we assign zero (czda=0 when the sun is below horizon) to an arbitrary negative value to avoid division by zero 
czda=xr.where(czda<=0,-0.5,czda)
````

plot the cosine zenith angle for the first time step

````python
cza[0,:,:].plot()
````

````python
czda[0,:,:].plot()
````

### 3.3 Calculate relative humidity
Here we calculate relative humidity from specific humidity since relative humidity is one of the required inputs of WBGT function.

````python
# calculate saturated vapor pressure
@vectorize
def esat(tas,ps):
    # tas: temperature (K)
    # ps: surface pressure (Pa)
    if tas>273.15:
        es = 611.21 * np.exp(17.502 * (tas - 273.15) *((tas - 32.18)**(-1)))
        es = (1.0007 + (3.46*10**(-6) * ps/100)) * es
    else:
        es = 611.15 * np.exp(22.452 * (tas - 273.15) * ((tas - 0.6)**(-1)))
        es=(1.0003 + (4.18*10**(-6) * ps/100)) * es
    return es
# calculate vapor pressure
def vaporpres(huss, ps):
    # huss: specific humidity (kg/kg)
    # ps: surface pressure (Pa)
    r=huss*((1-huss)**(-1))
    return ps*r*((0.622+r)**(-1))
# calculate relative humidity from specific humidity
def huss2rh(tas,huss,ps):
    # tas: temperature (K)
    # huss: specific humidity (kg/kg)
    # ps: surface pressure (Pa)
    return vaporpres(huss, ps)*(esat(tas,ps)**(-1))*100
````

````python
# calculate relative humidity
hurs=xr.apply_ufunc(huss2rh,tas,huss,ps)
# set the maximum value as 100%
hurs=xr.where(hurs>100,100,hurs)
````

### 3.4 Calculate wind speed
Calculate wind speed from the u and v components of wind speed. Keep in mind that the wind speed are output at 10m height.

````python
def xrsqrt(a, b):
    func = lambda x, y: np.sqrt(x ** 2 + y ** 2)
    return xr.apply_ufunc(func, a, b)
sfcwind=xrsqrt(uas,vas)
````

### 3.5 Calculate the ratio of direct solar radiation
We need to know the ratio of direct solar radiation in order to get the radiation heat gain right for WBGT calculation, since the treatment of direct and diffuse solar radiation are different in the code. In this case, we have ```rsdsdiff``` which is the diffuse solar radiation, so we can directly calculate the ratio of direct solar radiation. When we don't have such a field, we need to call the ```fdir``` function (also can be imported from the ```WBGT``` module) to calculate the ratio.

````python
f=(rsdsinterp-rsdsdiffinterp)/rsdsinterp 
# the treatments below aim to avoid unrealistically high or low values of f which are also included in Liljegren's code.
f=xr.where(cza<=np.cos(89.5/180*np.pi),0,f) 
f=xr.where(f>0.9,0.9,f)
f=xr.where(f<0,0,f)
f=xr.where(rsdsinterp<=0,0,f)
````

# 4. Calculate WBGT

### 4.1 Liljegren's original formulation
In Liljegren's original formulation, downward solar radiation is the only required radiation input, other radiation components are approximated internally.

The meaning of each argument:
- ```tas```: air temperature (K)
-  ```hurs```: relative humidity (%)
-  ```sfcwind```: 2 meter wind speed (m/s)
-  ```ps```: surface pressure (Pa)
-  ```rsdsinterp```: surface downward solar radiation (w/m2)
-  ```f```: the ratio of direct solar radiation 
- ```czda```: average cosine zenith angle during only the sunlit period of each interval

-  the ```False``` argument at the end tell the function that our wind speed is not at 2meter height which will make  the function to treat wind speed as 10 meter height and transfer it to 2 meter. The function currently only support 2 meter and 10meter wind. For wind speed at other heights, users need to change the source code slightly.

````python
wbgt_liljegren=xr.apply_ufunc(WBGT_Liljegren,tas,hurs,ps,sfcwind,rsdsinterp,f,czda,False)
wbgt_liljegren
````

We also exposed the functions for calculating natural wet bulb temperature and black globe temperature, two important components of WBGT

````python
from WBGT import Tg_Liljegren, Tnwb_Liljegren
````

````python
# calculate black globe temperature
tg_liljegren=xr.apply_ufunc(Tg_Liljegren,tas,hurs,ps,sfcwind,rsdsinterp,f,czda,False)
tg_liljegren
````

````python
# calculate natural wet bulb temperature
tnwb_liljegren=xr.apply_ufunc(Tnwb_Liljegren,tas,hurs,ps,sfcwind,rsdsinterp,f,czda,False)
tnwb_liljegren
````

### 4.2 Our modified formulation
We modify Liljegren's original code to take advantage of the full set of radiation components available from climate model output. 

The meaning of each argument:
- ```tas```: air temperature (K)
-  ```hurs```: relative humidity (%)
-  ```sfcwind```: 2 meter wind speed (m/s)
-  ```ps```: surface pressure (Pa)
-  ```rsdsinterp```: surface downward solar radiation (w/m2)
-  ```rsusinterp```: surface reflected solar radiation (w/m2)
-  ```rldsinterp```: surface downward long-wave radiation (w/m2)
-  ```rlusinterp```: surface upwelling long-wave radiation (w/m2)
- ```f```: the ratio of direct solar radiation 
-  ```czda```: average cosine zenith angle during only the sunlit period of each interval

````python
# calculate WBGT
wbgt_gcm=xr.apply_ufunc(WBGT_GCM,tas,hurs,ps,sfcwind,rsdsinterp,rsusinterp,rldsinterp,rlusinterp,f,czda,False)
wbgt_gcm
````

Let's plot the average WBGT at that day.

````python
wbgt_gcm.mean('time').plot()
````

calculate components of WBGT

````python
from WBGT import Tg_GCM, Tnwb_GCM
````

````python
# calculate black globe temperature
tg_gcm=xr.apply_ufunc(Tg_GCM,tas,ps,sfcwind,rsdsinterp,rsusinterp,rldsinterp,rlusinterp,f,czda,False)
tg_gcm
````

````python
# calculate natural wet bulb temperature
tnwb_gcm=xr.apply_ufunc(Tnwb_GCM,tas,hurs,ps,sfcwind,rsdsinterp,rsusinterp,rldsinterp,rlusinterp,f,czda,False)
tnwb_gcm
````
