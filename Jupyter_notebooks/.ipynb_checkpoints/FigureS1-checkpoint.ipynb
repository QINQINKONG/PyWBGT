# Treatment of cosine zenith angle

- The cosine zenith angle ($cos\theta$) is needed to project direct solar radiation from a flux through a horizontal plane (as stored in ERA5 reanalysis archive) to a flux through a plane perpendicular to the incoming solar radiation (as required by energy budget analysis). Since model radiation components are stored as accumulated-over-time quantities (over each hourly interval in the case of ERA5 reanalysis data), the time average of $cos\theta$ during each interval is needed. However, when the accumulation intervals encompass sunset or sunrise, the inclusion of zeros (when the sun is below the horizon) may make the time average of $cos\theta$ too small. Being in the denominator, this too small $cos\theta$ would lead to an overestimation of the projected direct solar radiation and consequently too high WBGT values. A simple approximate solution to this problem is taking the average $cos\theta$ during only the sunlit part of each interval.

- Here we compare WBGT values calculated from $cos\theta$ averaged during the whole interval and only the sunlit part of the interval through an example at a single grid cell (latitude: 36°, longitude: 275.75°, the location of Y-12 National Security Complex in U.S.) during year 2019.

- Please refer to Hogan and Hirahara (2016), or Di Napoli et al. (2020) for the calculation procedure of cos$\theta$ averaged during only the sunlit part of the interval. 

- In this notebook, we will reproduce Figure S1 in "Explicit Calculations of Wet Bulb Globe Temperature compared with approximations and why it matters for labor productivity" by Qinqin Kong and Matthew Huber.

````python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import ConnectionPatch
````

````python
%matplotlib inline
%config InlineBackend.figure_format = "retina"
````

### Read in data:
We read in the following files:
- ```WBGT_averge_cosz.nc``` which contains:
     - ```WBGT```: hourly WBGT during 2019 calculated from $cos\theta$ averaged during the whole interval
- ```WBGT_averge_cosz_sunlit.nc``` which contains:
     - ```WBGT```: hourly WBGT during 2019 calculated from $cos\theta$ averaged during only the sunlit period of each interval

````python
wbgt_average_cosz=xr.open_dataset('../nc_files/WBGT_averge_cosz.nc')
wbgt_average_cosz
````

````python
wbgt_average_cosz_sunlit=xr.open_dataset('../nc_files/WBGT_averge_cosz_sunlit.nc')
wbgt_average_cosz_sunlit
````

### Reproduce Figure S1

````python
# creat a figure
fig, (ax1,ax2) = plt.subplots(2,1,figsize=[10, 6])
# adjust the relative position of subplots
fig.subplots_adjust(wspace=0.1,hspace=0.3)

# plot the difference between WBGT values calculated from cosine zenith angle averaged 
# during the whole interval and only the sunlit part of each interval
(wbgt_average_cosz.WBGT-wbgt_average_cosz_sunlit.WBGT).plot(ax=ax1,color='black')
ax1.set_title('')
ax1.set_xlabel('')
ax1.set_ylim(-0.2,7)
ax1.xaxis.set_ticks_position('top')
ax1.set_ylabel('WBGT differences ($^\circ$C)',fontsize=12)
ax1.tick_params(axis='both', which='major', labelsize=12)

# zoom in a three day period from 2019-10-09 to 2019-10-11
(wbgt_average_cosz.WBGT-273.15).sel(time=slice('2019-10-9','2019-10-11')).plot(ax=ax2,color='red',linewidth=2)
(wbgt_average_cosz_sunlit.WBGT-273.15).sel(time=slice('2019-10-9','2019-10-11')).plot(ax=ax2,marker='x',linestyle='none',color='black')
ax2.set_title('')
ax2.set_xlabel('')
ax2.set_yticks([10,15,20,25])
ax2.set_ylim(10,28)
ax2.set_xlim(np.datetime64('2019-10-09'),np.datetime64('2019-10-12'))
ax2.set_xticks([np.datetime64('2019-10-09'),np.datetime64('2019-10-10'),np.datetime64('2019-10-11'),np.datetime64('2019-10-12')])
ax2.set_xticklabels(['10-09','10-10','10-11','10-12'], rotation = 0,horizontalalignment='center')
ax2.set_ylabel('WBGT ($^\circ$C)',fontsize=12)
ax2.tick_params(axis='both', which='major', labelsize=12)

xy1 = (np.datetime64('2019-10-09'),-0.2)
xy2 = (np.datetime64('2019-10-09'),28)
xy3 = (np.datetime64('2019-10-12'),-0.2)
xy4 = (np.datetime64('2019-10-12'),28)

con1 = ConnectionPatch(xyA=xy2, xyB=xy1, coordsA="data", coordsB='data', axesA=ax2, axesB=ax1, color="black")
ax2.add_artist(con1)
con1.set_linewidth(2)

con2 = ConnectionPatch(xyA=xy4, xyB=xy3, coordsA="data", coordsB='data', axesA=ax2, axesB=ax1, color="black")
ax2.add_artist(con2)
con2.set_linewidth(2)
````

### Figure caption:
Comparison between WBGTs calculated with the average cosine zenith angle during each hourly interval ($\overline{cos\theta}$) and that during only the sunlit part of each interval ($\overline{cos\theta_{sunlit}}$). A grid cell at 36$^\circ$N and $84.25^\circ$E (close to Y-12 National Security Complex in U.S.) is selected for demonstration purpose. Shown in upper panel are differences in hourly WBGT values calculated from two types of cosine zenith angle in 2019 (WBGT calculated from $\overline{cos\theta}$ - 
WBGT calculated from $\overline{cos\theta_{sunlit}}$). The lower panel zooms in a 3-day period from October 9 to 11, 2019 showing WBGTs calculated with $\overline{cos\theta}$ (red solid curve) and $\overline{cos\theta_{sunlit}}$ (black cross).

# Findings:
- WBGT values calculated from $\overline{cos\theta}$ show erroneous peaks around sunrise which can be removed by using $\overline{cos\theta_{sunlit}}$
